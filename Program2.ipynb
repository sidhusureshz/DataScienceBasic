```python
import numpy as np
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)
print("\nAddition:\n", A + B)
print("\nSubtraction:\n", A - B)
print("\nElement-wise Multiplication:\n", A * B)
print("\nMatrix Multiplication:\n", A @ B)
print("\nTranspose of A:\n", A.T)
print("\nSquare of A:\n", A ** 2)
print("\nRow Sum:\n", np.sum(A, axis=1))
print("\nColumn Sum:\n", np.sum(A, axis=0))
print("\nFlattened Matrix:")
print(A.flatten())
print("\nReshaped Matrix (1 x 4):")
print(A.reshape(1, 4))
print("\nRotated Matrix (90°):")
print(np.rot90(A))
print("\nFlipped Matrix:")
print(np.flipud(A))
M = np.array([[1, 2],
              [3, 4]])
print("\nMatrix for SVD:\n", M)
U, S, VT = np.linalg.svd(M)
print("\nU Matrix:\n", U)
print("\nSingular Values:\n", S)
print("\nVᵀ Matrix:\n", VT)
Sigma = np.diag(S)
Reconstructed = U @ Sigma @ VT
print("\nReconstructed Matrix:\n", Reconstructed)
```
